```python
do_grafu = df.filter(pl.col('real') == 'real').group_by(['kdo','kde']).len().sort(by="len",descending=True).join(
    df.filter(pl.col('real') != 'real').group_by(['kdo','kde']).len().sort(by="len",descending=True),
    how='left',
    on=['kdo','kde']
).fill_null(0).with_columns(
    (pl.col('len') + pl.col('len_right')).alias('celkem')
).rename(
    {'len':'skutečné','len_right':'halucinace'}
).filter(
    pl.col('celkem') >= 2
).sort(
    by="celkem",descending=True
).with_columns(
    pl.col("kdo").str.split(" ").list.slice(1,1)
).explode(
    "kdo"
).with_columns(
    pl.col("kdo").str.replace("Hynek","Mácha").str.replace("Jaromír","Erben")
).with_columns(
    (pl.col("kdo") + pl.lit(": ") + pl.col("kde")).alias("polozka")
).unpivot(
    index="polozka",on=["skutečné","halucinace"],variable_name="kolik"
).sort(
    by='kolik',descending=True
).with_columns(
    pl.col("kolik").str.replace("halucinace","vyhalucinované")
)

razeni = do_grafu.select(pl.col("polozka")).to_series().to_list()

do_grafu
```
